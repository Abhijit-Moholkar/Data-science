```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('company_sales.csv')
df
```

```python
# To drow univariate line plot

plt.plot(df.facecream,'r--o',label = 'Facecream')
plt.xticks(ticks = range(12),labels = df.month_number)
plt.xlabel('Month')
plt.ylabel('Sale Quantity')
plt.legend()
plt.show()
```

```python
plt.plot(df.facewash,'k-d',label = 'Facewash')
plt.xticks(ticks = range(12),labels = df.month_number)
plt.xlabel('Month')
plt.ylabel('Sale Quantity')
plt.legend()
plt.show()
```

```python
plt.plot(df.facecream,'r--o',label = 'Facecream')
plt.xticks(ticks = range(12),labels = df.month_number)

plt.plot(df.facewash,'k-d',label = 'Facewash')
plt.xticks(ticks = range(12),labels = df.month_number)
plt.xlabel('Month')
plt.ylabel('Sale Quantity')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (10,8))
plt.scatter(df.month_number,df.facewash,color = 'g',label = 'Facewash')
plt.scatter(df.month_number, df.facecream, color = 'r',label = 'Facecream')
plt.scatter(df.month_number, df.toothpaste,color = 'y',label = 'Toothpaste')
plt.scatter(df.month_number,df.bathingsoap,color = 'b',label = 'Bathing Soap')
plt.scatter(df.month_number,df.shampoo, color = 'm',label = 'Shampoo')
plt.scatter(df.month_number,df.moisturizer,color = 'k',label = 'Moisturizer')
plt.xlabel('Month')
plt.ylabel('Sale Quantity')
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (14,8))
plt.bar(df.month_number,df.facecream,label = 'Facecream',width = 0.4,align = 'edge',color = 'r')
plt.bar(df.month_number,df.facewash,label = 'Facewash',width = -0.4,align = 'edge',color = 'g')
plt.bar(df.month_number,df.toothpaste,label = 'Toothpaste',width = 0.3,align = 'edge',color = 'y')
plt.bar(df.month_number,df.shampoo,label = 'Shampoo',width = -0.3,align = 'edge',color = 'm')
plt.bar(df.month_number,df.moisturizer,label = 'Moisturizer',width = 0.5,align = 'edge',color = 'k')
plt.bar(df.month_number,df.bathingsoap,label = 'Bathingsoap',width = 0.2,align = 'center',color = 'b')
plt.xticks(ticks = range(1,13), labels = df.month_number)
plt.xlabel('Month')
plt.ylabel('Sale Quantity')
plt.legend()
plt.show()
```

```python
plt.hist(df.bathingsoap,bins = [1100,1500,2000,2500,3000,3500,4000,5000,6000,7000,8000,9000,10000,11000,12000,13000,14000,15000],rwidth = 0.2,color = 'b')
plt.hist(df.facecream,bins = [1100,1500,2000,2500,3000,3500,4000],rwidth = 0.4,color = 'r',align = 'right')
plt.hist(df.facewash,bins = [1100,1500,2000,2500,3000,3500,4000],rwidth = 0.3,color = 'g')
plt.show()
```
