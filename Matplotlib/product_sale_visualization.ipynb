# Visualizaton of beauty product sales data
It is to compare sale of beauty product

##### Import required libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

##### Read data from csv file

```python
df4=pd.read_csv('company_sales.csv')
df4
```

##### Visualize data distribution

```python
plt.scatter(df4.month_number,df4.total_units)
plt.show()
```

##### Compare sale of each beauty product

```python
plt.figure(figsize=(7,6))
plt.plot(df4.month_number,df4.facecream,'--ro',label='facecream')
plt.plot(df4.month_number,df4.facewash,label='facewash')
plt.plot(df4.month_number,df4.toothpaste,label='toothpaste')
plt.plot(df4.month_number,df4.bathingsoap,label='bathingsoap')
plt.plot(df4.month_number,df4.shampoo,label='shampoo')
plt.plot(df4.month_number,df4.moisturizer,label='moisturizer')
plt.savefig('line plot',dpi=300)
plt.xlabel('month')
plt.ylabel('Quantity')
plt.legend(loc=2)
plt.grid()
plt.show()
```

```python
# Visualize categorical data comparison 

plt.bar(df4.month_number,df4.facecream,width=0.3,align='edge',label='facecream')
plt.bar(df4.month_number,df4.facewash,width = -0.3,align='edge',label='facewash')
plt.xlabel('Month')
plt.ylabel('Sale')
plt.legend()
plt.show()
```

##### Visualize multiple plots together
It is to compare features in multiple perspectives

```python
plt.subplot(121)
plt.bar(df4.month_number,df4.facecream,width=0.3,align='edge',label='facecream')
plt.bar(df4.month_number,df4.facewash,width = -0.3,align='edge',label='facewash')
plt.xlabel('Month')
plt.ylabel('Sale')
plt.legend()
#plt.show()


plt.subplot(122)
plt.plot(df4.month_number,df4.facecream,'--ro',label='facecream')
plt.plot(df4.month_number,df4.facewash,label='facewash')
plt.plot(df4.month_number,df4.toothpaste,label='toothpaste')
plt.plot(df4.month_number,df4.bathingsoap,label='bathingsoap')
plt.plot(df4.month_number,df4.shampoo,label='shampoo')
plt.plot(df4.month_number,df4.moisturizer,label='moisturizer')
plt.savefig('line plot',dpi=300)
plt.xlabel('month')
plt.ylabel('Quantity')
plt.legend(loc=2)
plt.grid()
plt.tight_layout(h_pad=14)
plt.show()
```

```python
# Get total sale quantity of each product

item = ['facecream' ,'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
value = [sum(df4.facecream), sum(df4.facewash), sum(df4.toothpaste), sum(df4.bathingsoap), sum(df4.shampoo), sum(df4.moisturizer)] 
print(item)
print(value)
```

```python
# Visualize percentile distribution and highlight product with less sale

plt.pie(value,labels=item,autopct='%.f%%',explode=[0,0.3,0,0,0,0.3])
plt.show()
```
